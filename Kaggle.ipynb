```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:/Users/Dell/Desktop/train.csv')
df.head()
```

```python
df = df.set_index('PassengerId')
```

```python
df.isnull().sum()
```

```python
df = df.replace(r'\s+( +\.)|#',np.nan,regex=True).replace('',np.nan)
```

```python
df.isna().any()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df['Cabin'].fillna(method = 'ffill',inplace=True)
df['Embarked'].fillna(method = 'ffill',inplace=True)
```

```python
df['Cabin'].fillna(method = 'bfill',inplace=True)
```

```python
df1 = df.drop('Name', axis = 1)
df2 = df1.drop('Ticket', axis = 1)
df2.head()
```

```python
bins = [0, 25, 45, 80]
names = ['young adults', 'adults', 'seniors']
df2['Age_Bins'] = pd.cut(df2['Age'], bins, labels=names)
df3 = df2.drop('Age', axis = 1)
df3.head()
```

```python
df3.Cabin.unique()
```

```python
result = df3['Cabin'].str.split('([A-Za-z]+)(\d+)', expand=True)
result = result.loc[:,[1,2]]
result.rename(columns={1:'x', 2:'y'}, inplace=True)
df4 = result.drop('y', axis = 1)
df4.x.unique()
```

```python
df4.isna().any()
```

```python
df4['x'].fillna(method = 'ffill',inplace=True)
df4.x.unique()
```

```python
df3['Cab'] = df4['x']
```

```python
df3.head()
```

```python
df5 = df3.drop('Cabin', axis = 1)
```

```python
df5.head()
```

```python
df5.Embarked.unique()
```

```python
mapping = {'male': 0, 'female': 1}
df5.replace({'Sex': mapping})
mapping1 = {'S': 0, 'C': 1, 'Q': 2}
df5.replace({'Sex': mapping}).replace({'Embarked': mapping1})
mapping2 = {'young adults': 0, 'adults': 1, 'seniors': 2}
df5.replace({'Sex': mapping}).replace({'Embarked': mapping1}).replace({'Age_Bins': mapping2})
mapping3 = {'C': 0, 'E': 1, 'G': 2, 'D': 3, 'A': 4, 'B': 5, 'F': 6}
df5.replace({'Sex': mapping}).replace({'Embarked': mapping1}).replace({'Age_Bins': mapping2}).replace({'Cab': mapping3})
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df5.replace({'Sex': mapping}).replace({'Embarked': mapping1}).replace({'Age_Bins': mapping2}).replace({'Cab': mapping3}))
```

```python
new_df = pd.DataFrame(df_scaled, columns=df5.columns, index=df5.index)
new_df.head()
```

```python
X = new_df.drop('Survived', axis = 1)
Y = new_df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='lbfgs').fit(X_train, Y_train)
print('Accuracy of Logistic regression classifier on test set: {:.2f}'
     .format(clf.score(X_test, Y_test)))
```

```python
from sklearn.linear_model import Ridge
logridge = Ridge(alpha=20.0).fit(X_train, Y_train)
print('R-squared score (test): {:.3f}'
     .format(logridge.score(X_test, Y_test)))
print('Number of non-zero features: {}'
     .format(np.sum(logridge.coef_ != 0)))
```

```python
from sklearn.neural_network import MLPClassifier
clf2 = MLPClassifier(hidden_layer_sizes = [100, 100], alpha = 5.0,
                   random_state = 0, solver='lbfgs').fit(X_train, Y_train)
print('Accuracy of NN classifier on test set: {:.2f}'
     .format(clf2.score(X_test, Y_test)))
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
clf3 = tree.DecisionTreeClassifier(criterion='gini', min_samples_split=2, max_depth=None)
clf3 = clf3.fit(X_train, Y_train)
score = clf3.score(X_test, Y_test)
print(score)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf4 = RandomForestClassifier() 
clf4.fit(X_train, Y_train)
score = clf4.score(X_test, Y_test)
print('Accuracy:', score)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
grad=GradientBoostingClassifier()
grad.fit(X_train,Y_train)
score = grad.score(X_test, Y_test)
print(score)
```

```python
from sklearn import svm
clf5 = svm.SVC(kernel='rbf')
clf5.fit(X_train, Y_train)
score=clf5.score(X_test, Y_test)
print(score)
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.layers.core import Dropout
from keras.callbacks import EarlyStopping
from keras.layers import BatchNormalization
from keras.initializers import Zeros, RandomNormal, he_normal
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(BatchNormalization())
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1, activation='sigmoid'))

#Compile the model:
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])

#Fit the model:
model.fit(X_train, Y_train, batch_size=32, epochs=3)

#Evaluate the model:
test_loss, test_acc = model.evaluate(X_test,  Y_test, verbose=2)
print('\nTest loss:', test_loss)
print('\nTest accuracy:', test_acc)
```
